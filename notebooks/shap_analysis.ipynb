```python
import shap
import joblib
import pandas as pd

# Load trained model
model = joblib.load("../models/frustration_model.pkl")

# Load dataset
data = pd.read_csv("../data/raw/user_behavior.csv")

# Separate features
X = data.drop("frustrated", axis=1)

print("Model and data loaded successfully!")
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer(X)

print("SHAP values generated!")
```

```python
shap.summary_plot(shap_values.values, X)
```

```python
shap.plots.waterfall(shap_values[0, :, 1])
```
